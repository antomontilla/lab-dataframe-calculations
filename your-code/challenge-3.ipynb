## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# Sub Problem 1: How to aggregate the  amount_spent for unique customers?
orders_by_customer = orders.groupby(['CustomerID']).sum()
orders_by_customer = orders_by_customer.reset_index()
orders_by_customer
```

```python
# Sub Problem 2 How to select customers whose aggregated amount_spent is in a given quantile range?
display(orders_by_customer['amount_spent'].min())
display(orders_by_customer['amount_spent'].max())
display(orders_by_customer['amount_spent'].mean())
display(orders_by_customer['amount_spent'].median())
display(orders_by_customer['amount_spent'].quantile(0.75))

#I could use any of these references to filter customers based on their amount spent:
orders_by_customer[orders_by_customer['amount_spent']>orders_by_customer['amount_spent'].quantile(0.75)]

```

```python
# Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?¶
# A reasonable assumption could be to label customers as "VIP" if amount spent is >= than the 0.75 quantile
# I will also label customers as "Preferred" if amount spent is more than the average but < than the 0.75 quantile
# I will label other customers as "Other"

binnames = ["Other", "Preferred", "VIP"]
bins = pd.cut(orders_by_customer['amount_spent'],[-0.1,674.45,1661.64,280207], labels = binnames)
bins.value_counts(dropna=False)
```

```python
#Will create a function to add these labels into the dataframe

def client_label(value):
    if value > 1661.64:
        return "VIP"
    elif (value >= 674.45) & (value < 1661.64):
        return "Preferred"
    else:
        return "Other"

orders_by_customer['client label'] = orders_by_customer['amount_spent'].apply(client_label)
orders_by_customer.head(30)
#selecting only the columns 'CustomerID' & 'client label' into a new table
client_label_table = orders_by_customer[['CustomerID','client label']]
client_label_table
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
# your code here
#First I merge the orders table with the client_label table so it adds the label of the client
orders2 = pd.merge(left = orders,
                                 right = client_label_table,
                                 how = 'inner', 
                                 left_on = "CustomerID", 
                                 right_on= "CustomerID")
orders2.head()
#Now I will filter this orders2 df with customers labelled as VIP and group by country and customerID:
vip_customers_by_country = orders2[orders2['client label'] == 'VIP'].groupby('Country')['CustomerID'].nunique()
vip_customers_by_country
vip_customers_by_country.idxmax() #United Kingdom has the most VIP customers
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
#Now I will filter this orders2 df with customers labelled as VIP and group by country and customerID:
vip_pre_customers_by_country = orders2[(orders2['client label'] == 'VIP') | (orders2['client label'] == 'Preferred')].groupby('Country')['CustomerID'].nunique()
vip_pre_customers_by_country
vip_pre_customers_by_country.idxmax() #United Kingdom has the most VIP and preferred customers
```
